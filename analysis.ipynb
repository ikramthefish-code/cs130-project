```python
#making a function that looks at all of the dairy 
def milk():
    return food[(food["Category"].str.contains("Milk"))| 
                    food["Category"].str.contains("milk") |  food["Category"].str.contains("Cheese") | food["Category"].str.contains("Yogurt") | (food["Description"].str.contains("milk")) & (food["Category"] != "Grits") & (food["Category"] != "Rice") & (food["Category"] != "Oatmeal") ]
milk()
```

```python
#making a Function looking for Things High in Carb
def Carbs():
    newdata = food.drop_duplictes(subset = ["Category"], keep = "first")
    return newdata.sort_values(by ="Carbohydrate")
Carbs()
```
